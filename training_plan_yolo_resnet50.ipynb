## 1. ResNet Dataset Audit

Checks folders, image readability, dimensions, RGB compatibility, low-quality samples, exact duplicates.


````python
def iter_images(root):
    root = Path(root)
    if not root.exists():
        return
    for p in root.rglob("*"):
        if p.is_file() and p.suffix.lower() in IMG_EXTS:
            yield p

def image_sha256(path):
    h = hashlib.sha256()
    with open(path, "rb") as f:
        for chunk in iter(lambda: f.read(1024 * 1024), b""):
            h.update(chunk)
    return h.hexdigest()

def audit_resnet(root=RESNET_ROOT):
    rows, bad, structure = [], [], []
    for split in ["train", "valid", "test"]:
        split_dir = root / split
        if not split_dir.exists():
            structure.append({"path": str(split_dir), "issue": "missing_split"})
        for cls in CLASSES:
            d = root / split / cls
            if not d.exists():
                structure.append({"path": str(d), "issue": "missing_class_folder"})
                continue
            for p in iter_images(d):
                try:
                    with Image.open(p) as im:
                        im.verify()
                    with Image.open(p) as im:
                        rgb = im.convert("RGB")
                        stat = ImageStat.Stat(rgb)
                        rows.append({
                            "path": str(p),
                            "rel": str(p.relative_to(root)),
                            "split": split,
                            "class": cls,
                            "width": im.width,
                            "height": im.height,
                            "mode": im.mode,
                            "bytes": p.stat().st_size,
                            "mean": float(np.mean(stat.mean)),
                            "std": float(np.mean(stat.stddev)),
                            "sha256": image_sha256(p),
                        })
                except Exception as e:
                    bad.append({"path": str(p), "issue": repr(e)})
    df = pd.DataFrame(rows)
    bad_df = pd.DataFrame(bad)
    struct_df = pd.DataFrame(structure)
    df.to_csv(REPORTS / "notebook_resnet_audit.csv", index=False)
    bad_df.to_csv(REPORTS / "notebook_resnet_bad_files.csv", index=False)
    struct_df.to_csv(REPORTS / "notebook_resnet_structure_issues.csv", index=False)
    return df, bad_df, struct_df

resnet_df, bad_df, struct_df = audit_resnet()
print("images", len(resnet_df), "bad", len(bad_df), "structure issues", len(struct_df))
resnet_df.head()
````

## 3. Optional ResNet 70/15/15 Re-split

Dry-run by default. Set `APPLY_RESPLIT = True` only when ready.

Creates backup: `dataset/resnet_before_70_15_15_backup`.


````python
def stratified_resplit(root=RESNET_ROOT, apply=False):
    root = Path(root)
    backup = root.parent / "resnet_before_70_15_15_backup"
    plan = []
    files_by_class = {}
    for cls in CLASSES:
        files = []
        for split in ["train", "valid", "test"]:
            files.extend(list(iter_images(root / split / cls)))
        files = sorted(files)
        random.Random(SEED).shuffle(files)
        files_by_class[cls] = files
        n = len(files)
        n_train = int(round(n * 0.70))
        n_valid = int(round(n * 0.15))
        n_test = n - n_train - n_valid
        plan.append({"class": cls, "total": n, "train": n_train, "valid": n_valid, "test": n_test})
    plan_df = pd.DataFrame(plan)
    display(plan_df)
    if not apply:
        print("DRY RUN. Set APPLY_RESPLIT=True to rewrite dataset/resnet splits.")
        return plan_df
    if backup.exists():
        raise RuntimeError(f"Backup already exists: {backup}")
    shutil.copytree(root, backup)
    for split in ["train", "valid", "test"]:
        for cls in CLASSES:
            d = root / split / cls
            d.mkdir(parents=True, exist_ok=True)
            for p in d.iterdir():
                if p.is_file() and p.suffix.lower() in IMG_EXTS:
                    p.unlink()
    for cls, files in files_by_class.items():
        n = len(files)
        n_train = int(round(n * 0.70))
        n_valid = int(round(n * 0.15))
        parts = {
            "train": files[:n_train],
            "valid": files[n_train:n_train + n_valid],
            "test": files[n_train + n_valid:],
        }
        for split, split_files in parts.items():
            for i, src in enumerate(split_files):
                dst = root / split / cls / f"{cls}_{i:05d}{src.suffix.lower()}"
                shutil.copy2(src, dst)
    print("re-split complete; backup:", backup)
    return plan_df

APPLY_RESPLIT = False
resplit_plan = stratified_resplit(apply=APPLY_RESPLIT)
````

## 4. YOLO Dataset Audit

Checks image-label pairing and YOLO box format.


````python
def audit_yolo(root=YOLO_ROOT):
    rows = []
    for split in ["train", "valid", "test"]:
        img_dir = root / split / "images"
        lbl_dir = root / split / "labels"
        for img in iter_images(img_dir):
            rel = img.relative_to(img_dir)
            label = (lbl_dir / rel).with_suffix(".txt")
            issues, boxes = [], 0
            if not label.exists():
                issues.append("missing_label")
            else:
                for line in label.read_text(encoding="utf-8", errors="ignore").splitlines():
                    parts = line.strip().split()
                    if not parts:
                        continue
                    if len(parts) != 5:
                        issues.append("bad_format")
                        continue
                    try:
                        cls = int(float(parts[0]))
                        vals = [float(x) for x in parts[1:]]
                        if cls < 0 or any(v < 0 or v > 1 for v in vals):
                            issues.append("bad_values")
                        else:
                            boxes += 1
                    except Exception:
                        issues.append("bad_values")
            try:
                with Image.open(img) as im:
                    w, h = im.size
            except Exception:
                w, h = None, None
                issues.append("unreadable_image")
            rows.append({"split": split, "image": str(img), "label": str(label), "width": w, "height": h, "boxes": boxes, "issue": ";".join(sorted(set(issues)))})
    df = pd.DataFrame(rows)
    df.to_csv(REPORTS / "notebook_yolo_audit.csv", index=False)
    return df

yolo_df = audit_yolo()
display(yolo_df.groupby("split").agg(images=("image", "count"), boxes=("boxes", "sum"), issues=("issue", lambda s: (s != "").sum())))
yolo_df[yolo_df.issue != ""].head(30)
````

## 6. Advanced ResNet50 Classifier Training

Research-safe upgrades added before final training:

- **MixUp**: training-loop augmentation only; blends images/labels to reduce overconfidence and improve boundary smoothness.
- **CutMix**: training-loop augmentation only; replaces image patches between samples and mixes labels by visible area.
- **Hybrid policy**: each training batch chooses CutMix with probability `CUTMIX_PROB`; otherwise MixUp may run with probability `MIXUP_PROB`.
- **Label smoothing**: included inside soft-target Focal Loss to improve calibration and reduce noisy-label sensitivity.
- **EMA weights**: validation and final test use exponential moving average model weights for better generalization.
- **AdamW + weight decay + gradient clipping**: stable fine-tuning with controlled update size.
- **AMP mixed precision**: enabled on CUDA for speed and regularization-like numerical noise.
- **CosineAnnealingWarmRestarts**: smoother LR schedule than plateau-only for augmentation-heavy fine-tuning.
- **5-fold Stratified CV**: uses `train+valid` only; `test` remains untouched until final evaluation.

Expected impact: MixUp/CutMix/label smoothing usually improve macro-F1 and minority recall (`scab`, `greening`) by reducing overfit to majority-class texture shortcuts. EMA/TTA often improve robustness and calibration. Accuracy may stay similar or slightly increase; macro-F1 is the main selection metric.


````python
import copy, json, math, random, time
from pathlib import Path

import numpy as np
import pandas as pd
import torch
from PIL import Image
from torch import nn
from torch.utils.data import DataLoader, Dataset, Subset, WeightedRandomSampler
from torchvision import datasets, transforms, models
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score, balanced_accuracy_score,
    classification_report, confusion_matrix, roc_auc_score
)
from sklearn.preprocessing import label_binarize

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
USE_AMP = torch.cuda.is_available()
SEED = 42
BATCH_SIZE = 32
NUM_WORKERS = 0
N_FOLDS = 5

# Advanced augmentation/loss knobs.
MIXUP_ALPHA = 0.30
MIXUP_PROB = 0.50
CUTMIX_ALPHA = 1.00
CUTMIX_PROB = 0.50
LABEL_SMOOTHING = 0.05
FOCAL_GAMMA = 2.0
EMA_DECAY = 0.999
GRAD_CLIP_NORM = 1.0
WEIGHT_DECAY = 5e-5
LR_HEAD = 1e-3
LR_FINETUNE = 3e-5
EPOCHS_HEAD = 12
EPOCHS_FINETUNE = 30
PATIENCE = 8
TTA_STEPS = 8

IMAGENET_MEAN = [0.485, 0.456, 0.406]
IMAGENET_STD = [0.229, 0.224, 0.225]

random.seed(SEED)
np.random.seed(SEED)
torch.manual_seed(SEED)
torch.cuda.manual_seed_all(SEED)
torch.backends.cudnn.benchmark = False
torch.backends.cudnn.deterministic = True

train_tfms = transforms.Compose([
    transforms.RandomResizedCrop(224, scale=(0.70, 1.0), ratio=(0.80, 1.25)),
    transforms.RandomHorizontalFlip(p=0.5),
    transforms.RandomVerticalFlip(p=0.25),
    transforms.RandomRotation(25),
    transforms.ColorJitter(brightness=0.25, contrast=0.25, saturation=0.20, hue=0.04),
    transforms.RandomAffine(degrees=0, translate=(0.06, 0.06), scale=(0.85, 1.15), shear=8),
    transforms.RandomPerspective(distortion_scale=0.20, p=0.25),
    transforms.GaussianBlur(kernel_size=3, sigma=(0.1, 1.2)),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
    transforms.RandomErasing(p=0.20, scale=(0.02, 0.12), ratio=(0.3, 3.3), value="random"),
])

eval_tfms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

tta_tfms = [
    eval_tfms,
    transforms.Compose([transforms.Resize((224, 224)), transforms.RandomHorizontalFlip(p=1.0), transforms.ToTensor(), transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD)]),
    transforms.Compose([transforms.Resize((224, 224)), transforms.RandomVerticalFlip(p=1.0), transforms.ToTensor(), transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD)]),
    transforms.Compose([transforms.Resize((224, 224)), transforms.RandomRotation((10, 10)), transforms.ToTensor(), transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD)]),
    transforms.Compose([transforms.Resize((224, 224)), transforms.RandomRotation((-10, -10)), transforms.ToTensor(), transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD)]),
    transforms.Compose([transforms.Resize((224, 224)), transforms.ColorJitter(brightness=0.10, contrast=0.10, saturation=0.08), transforms.ToTensor(), transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD)]),
    transforms.Compose([transforms.Resize((224, 224)), transforms.RandomAffine(degrees=0, translate=(0.03, 0.03), scale=(0.95, 1.05)), transforms.ToTensor(), transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD)]),
    transforms.Compose([transforms.Resize((224, 224)), transforms.RandomPerspective(distortion_scale=0.10, p=1.0), transforms.ToTensor(), transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD)]),
]

class TransformSubset(Dataset):
    def __init__(self, base_ds, indices, transform):
        self.base_ds = base_ds
        self.indices = list(indices)
        self.transform = transform
        self.classes = base_ds.classes
        self.class_to_idx = base_ds.class_to_idx
        self.samples = [base_ds.samples[i] for i in self.indices]
        self.targets = [base_ds.targets[i] for i in self.indices]
    def __len__(self):
        return len(self.indices)
    def __getitem__(self, idx):
        path, y = self.samples[idx]
        img = Image.open(path).convert("RGB")
        return self.transform(img), y

def build_cv_base_dataset():
    train_base = datasets.ImageFolder(RESNET_ROOT / "train")
    valid_base = datasets.ImageFolder(RESNET_ROOT / "valid")
    if train_base.class_to_idx != valid_base.class_to_idx:
        raise RuntimeError(f"class mapping mismatch: {train_base.class_to_idx} vs {valid_base.class_to_idx}")
    train_base.samples = list(train_base.samples) + list(valid_base.samples)
    train_base.imgs = train_base.samples
    train_base.targets = [y for _p, y in train_base.samples]
    return train_base

cv_base_ds = build_cv_base_dataset()
train_ds = datasets.ImageFolder(RESNET_ROOT / "train", transform=train_tfms)   # kept for downstream class names
valid_ds = datasets.ImageFolder(RESNET_ROOT / "valid", transform=eval_tfms)    # deterministic eval only
test_ds = datasets.ImageFolder(RESNET_ROOT / "test", transform=eval_tfms)      # untouched held-out test

print("device", DEVICE)
print("classes", cv_base_ds.class_to_idx)
print("train+valid", len(cv_base_ds), "test", len(test_ds))
print("counts train+valid", np.bincount(cv_base_ds.targets, minlength=len(cv_base_ds.classes)).tolist())
````

````python
def one_hot(y, num_classes, smoothing=0.0):
    off = smoothing / num_classes
    on = 1.0 - smoothing + off
    out = torch.full((y.size(0), num_classes), off, device=y.device)
    out.scatter_(1, y.view(-1, 1), on)
    return out

class SoftTargetFocalLoss(nn.Module):
    """Focal loss supporting hard or soft targets, with label smoothing compatibility."""
    def __init__(self, gamma=2.0, class_weight=None, label_smoothing=0.0):
        super().__init__()
        self.gamma = gamma
        self.label_smoothing = label_smoothing
        self.register_buffer("class_weight", class_weight if class_weight is not None else None)
    def forward(self, logits, target):
        num_classes = logits.size(1)
        log_prob = torch.log_softmax(logits, dim=1)
        prob = log_prob.exp()
        if target.dtype in (torch.long, torch.int64):
            target = one_hot(target, num_classes, self.label_smoothing)
        elif self.label_smoothing > 0:
            target = target * (1.0 - self.label_smoothing) + self.label_smoothing / num_classes
        focal = (1.0 - prob).pow(self.gamma)
        loss = -target * focal * log_prob
        if self.class_weight is not None:
            loss = loss * self.class_weight.view(1, -1)
        return loss.sum(dim=1).mean()

def class_weight_from_targets(targets, num_classes):
    counts = np.bincount(np.array(targets), minlength=num_classes).astype(np.float32)
    counts[counts == 0] = 1.0
    inv = 1.0 / counts
    weights = inv / inv.sum() * num_classes
    return torch.tensor(weights, dtype=torch.float32, device=DEVICE)

def make_weighted_sampler(targets, num_classes):
    counts = np.bincount(np.array(targets), minlength=num_classes).astype(np.float64)
    counts[counts == 0] = 1.0
    class_weights = 1.0 / counts
    sample_weights = torch.DoubleTensor([class_weights[y] for y in targets])
    return WeightedRandomSampler(sample_weights, num_samples=len(sample_weights), replacement=True)

def mixup_batch(x, y, num_classes, alpha=MIXUP_ALPHA):
    lam = np.random.beta(alpha, alpha) if alpha > 0 else 1.0
    perm = torch.randperm(x.size(0), device=x.device)
    mixed_x = lam * x + (1.0 - lam) * x[perm]
    y_a = one_hot(y, num_classes, LABEL_SMOOTHING)
    y_b = one_hot(y[perm], num_classes, LABEL_SMOOTHING)
    mixed_y = lam * y_a + (1.0 - lam) * y_b
    return mixed_x, mixed_y

def rand_bbox(size, lam):
    _, _, h, w = size
    cut_ratio = math.sqrt(1.0 - lam)
    cut_w = int(w * cut_ratio)
    cut_h = int(h * cut_ratio)
    cx = np.random.randint(w)
    cy = np.random.randint(h)
    x1 = np.clip(cx - cut_w // 2, 0, w)
    y1 = np.clip(cy - cut_h // 2, 0, h)
    x2 = np.clip(cx + cut_w // 2, 0, w)
    y2 = np.clip(cy + cut_h // 2, 0, h)
    return x1, y1, x2, y2

def cutmix_batch(x, y, num_classes, alpha=CUTMIX_ALPHA):
    lam = np.random.beta(alpha, alpha) if alpha > 0 else 1.0
    perm = torch.randperm(x.size(0), device=x.device)
    x1, y1, x2, y2 = rand_bbox(x.size(), lam)
    mixed_x = x.clone()
    mixed_x[:, :, y1:y2, x1:x2] = x[perm, :, y1:y2, x1:x2]
    lam = 1.0 - ((x2 - x1) * (y2 - y1) / (x.size(-1) * x.size(-2)))
    y_a = one_hot(y, num_classes, LABEL_SMOOTHING)
    y_b = one_hot(y[perm], num_classes, LABEL_SMOOTHING)
    mixed_y = lam * y_a + (1.0 - lam) * y_b
    return mixed_x, mixed_y

def apply_mix_strategy(x, y, num_classes):
    # Insertion point: MixUp/CutMix are applied here, inside the training loop only.
    r = random.random()
    if r < CUTMIX_PROB:
        return cutmix_batch(x, y, num_classes)
    if r < CUTMIX_PROB + MIXUP_PROB:
        return mixup_batch(x, y, num_classes)
    return x, one_hot(y, num_classes, LABEL_SMOOTHING)

class ModelEMA:
    # Insertion point: create once per fold; call update() after every optimizer step.
    def __init__(self, model, decay=EMA_DECAY):
        self.ema = copy.deepcopy(model).eval()
        self.decay = decay
        for p in self.ema.parameters():
            p.requires_grad_(False)
    @torch.no_grad()
    def update(self, model):
        msd = model.state_dict()
        for k, v in self.ema.state_dict().items():
            if v.dtype.is_floating_point:
                v.copy_(v * self.decay + msd[k].detach() * (1.0 - self.decay))
            else:
                v.copy_(msd[k])

def build_resnet50(num_classes, freeze=True):
    weights = models.ResNet50_Weights.IMAGENET1K_V2
    model = models.resnet50(weights=weights)
    if freeze:
        for p in model.parameters():
            p.requires_grad = False
    model.fc = nn.Sequential(nn.Dropout(0.35), nn.Linear(model.fc.in_features, num_classes))
    return model.to(DEVICE)

def set_phase(model, phase):
    for p in model.parameters():
        p.requires_grad = False
    if phase == "head":
        for p in model.fc.parameters():
            p.requires_grad = True
    elif phase == "finetune":
        # BatchNorm kept in eval mode during fine-tune to avoid unstable stats on small minority folds.
        for name, p in model.named_parameters():
            if name.startswith("layer4") or name.startswith("fc"):
                p.requires_grad = True
    else:
        raise ValueError(phase)

def freeze_batchnorm_stats(model):
    for m in model.modules():
        if isinstance(m, nn.BatchNorm2d):
            m.eval()
````

````python
def compute_metrics(y_true, y_prob, classes):
    y_pred = y_prob.argmax(axis=1)
    out = {
        "accuracy": accuracy_score(y_true, y_pred),
        "precision_macro": precision_score(y_true, y_pred, average="macro", zero_division=0),
        "recall_macro": recall_score(y_true, y_pred, average="macro", zero_division=0),
        "macro_f1": f1_score(y_true, y_pred, average="macro", zero_division=0),
        "balanced_accuracy": balanced_accuracy_score(y_true, y_pred),
        "confusion_matrix": confusion_matrix(y_true, y_pred).tolist(),
        "classification_report": classification_report(y_true, y_pred, target_names=classes, digits=4, zero_division=0),
    }
    try:
        if len(np.unique(y_true)) == len(classes):
            y_bin = label_binarize(y_true, classes=list(range(len(classes))))
            out["roc_auc_ovr_macro"] = roc_auc_score(y_bin, y_prob, average="macro", multi_class="ovr")
        else:
            out["roc_auc_ovr_macro"] = None
    except ValueError:
        out["roc_auc_ovr_macro"] = None
    return out

@torch.no_grad()
def predict_loader(model, loader):
    model.eval()
    y_true, y_prob = [], []
    for x, y in loader:
        logits = model(x.to(DEVICE))
        prob = torch.softmax(logits, dim=1).cpu().numpy()
        y_true.extend(y.numpy().tolist())
        y_prob.extend(prob)
    return np.array(y_true), np.array(y_prob)

@torch.no_grad()
def predict_tta(model, samples, labels, max_tta=TTA_STEPS):
    model.eval()
    tfms = tta_tfms[:max(1, max_tta)]
    probs = []
    for path, _ in samples:
        img = Image.open(path).convert("RGB")
        batch = torch.stack([tfm(img) for tfm in tfms]).to(DEVICE)
        logits = model(batch)
        probs.append(torch.softmax(logits, dim=1).mean(dim=0).cpu().numpy())
    return np.array(labels), np.array(probs)

def train_one_epoch(model, ema, loader, criterion, optimizer, scaler, num_classes):
    model.train()
    freeze_batchnorm_stats(model)
    total_loss, total_n = 0.0, 0
    for x, y in loader:
        x, y = x.to(DEVICE), y.to(DEVICE)
        x, soft_y = apply_mix_strategy(x, y, num_classes)  # MixUp/CutMix insertion point.
        optimizer.zero_grad(set_to_none=True)
        with torch.cuda.amp.autocast(enabled=USE_AMP):
            logits = model(x)
            loss = criterion(logits, soft_y)
        scaler.scale(loss).backward()
        scaler.unscale_(optimizer)
        torch.nn.utils.clip_grad_norm_(model.parameters(), GRAD_CLIP_NORM)
        scaler.step(optimizer)
        scaler.update()
        ema.update(model)  # EMA insertion point.
        total_loss += float(loss.item()) * x.size(0)
        total_n += x.size(0)
    return total_loss / max(total_n, 1)

def train_phase(model, ema, phase, train_loader, valid_loader, criterion, lr, epochs, best, fold_dir, start_epoch=0):
    set_phase(model, phase)
    freeze_batchnorm_stats(model)
    optimizer = torch.optim.AdamW([p for p in model.parameters() if p.requires_grad], lr=lr, weight_decay=WEIGHT_DECAY)
    scheduler = torch.optim.lr_scheduler.CosineAnnealingWarmRestarts(optimizer, T_0=max(4, epochs // 3), T_mult=1)
    scaler = torch.cuda.amp.GradScaler(enabled=USE_AMP)
    history, bad_epochs = [], 0
    for ep in range(1, epochs + 1):
        epoch = start_epoch + ep
        tr_loss = train_one_epoch(model, ema, train_loader, criterion, optimizer, scaler, len(train_loader.dataset.classes))
        scheduler.step(ep)
        y_true, y_prob = predict_loader(ema.ema, valid_loader)  # Validate EMA weights.
        va_metrics = compute_metrics(y_true, y_prob, valid_loader.dataset.classes)
        va_loss = np.nan
        row = {
            "epoch": epoch, "phase": phase, "train_loss": tr_loss,
            "valid_macro_f1": va_metrics["macro_f1"],
            "valid_balanced_accuracy": va_metrics["balanced_accuracy"],
            "valid_accuracy": va_metrics["accuracy"],
            "lr": optimizer.param_groups[0]["lr"],
        }
        history.append(row)
        print(row)
        improved = va_metrics["macro_f1"] > best["macro_f1"] + 1e-5
        if improved:
            best.update({"macro_f1": va_metrics["macro_f1"], "epoch": epoch, "phase": phase})
            torch.save(ema.ema.state_dict(), fold_dir / "resnet50_ema_best.pt")
            (fold_dir / "valid_metrics_best.json").write_text(json.dumps(va_metrics, indent=2), encoding="utf-8")
            bad_epochs = 0
        else:
            bad_epochs += 1
            if bad_epochs >= PATIENCE:
                print(f"early stopping fold phase={phase} epoch={epoch}")
                break
    return history, best
````

````python
def train_resnet50_stratified_cv():
    out_dir = ROOT / "runs" / "resnet50_cv_advanced" / time.strftime("%Y%m%d_%H%M%S")
    out_dir.mkdir(parents=True, exist_ok=True)
    targets = np.array(cv_base_ds.targets)
    skf = StratifiedKFold(n_splits=N_FOLDS, shuffle=True, random_state=SEED)
    fold_rows = []

    for fold, (train_idx, valid_idx) in enumerate(skf.split(np.zeros(len(targets)), targets), start=1):
        print(f"\n===== FOLD {fold}/{N_FOLDS} =====")
        fold_dir = out_dir / f"fold_{fold}"
        fold_dir.mkdir(parents=True, exist_ok=True)

        tr_ds = TransformSubset(cv_base_ds, train_idx, train_tfms)
        va_ds = TransformSubset(cv_base_ds, valid_idx, eval_tfms)  # No validation augmentation.
        sampler = make_weighted_sampler(tr_ds.targets, len(cv_base_ds.classes))
        tr_loader = DataLoader(tr_ds, batch_size=BATCH_SIZE, sampler=sampler, num_workers=NUM_WORKERS, pin_memory=torch.cuda.is_available())
        va_loader = DataLoader(va_ds, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=torch.cuda.is_available())

        class_w = class_weight_from_targets(tr_ds.targets, len(cv_base_ds.classes))
        criterion = SoftTargetFocalLoss(gamma=FOCAL_GAMMA, class_weight=class_w, label_smoothing=LABEL_SMOOTHING)
        model = build_resnet50(len(cv_base_ds.classes), freeze=True)
        ema = ModelEMA(model, decay=EMA_DECAY)
        best = {"macro_f1": -1.0, "epoch": 0, "phase": "none"}
        history = []

        h, best = train_phase(model, ema, "head", tr_loader, va_loader, criterion, LR_HEAD, EPOCHS_HEAD, best, fold_dir, start_epoch=0)
        history.extend(h)
        if (fold_dir / "resnet50_ema_best.pt").exists():
            model.load_state_dict(torch.load(fold_dir / "resnet50_ema_best.pt", map_location=DEVICE))
            ema = ModelEMA(model, decay=EMA_DECAY)
        h, best = train_phase(model, ema, "finetune", tr_loader, va_loader, criterion, LR_FINETUNE, EPOCHS_FINETUNE, best, fold_dir, start_epoch=EPOCHS_HEAD)
        history.extend(h)
        pd.DataFrame(history).to_csv(fold_dir / "history.csv", index=False)

        best_model = build_resnet50(len(cv_base_ds.classes), freeze=False)
        best_model.load_state_dict(torch.load(fold_dir / "resnet50_ema_best.pt", map_location=DEVICE))
        y_true, y_prob = predict_loader(best_model, va_loader)
        metrics = compute_metrics(y_true, y_prob, cv_base_ds.classes)
        print(metrics["classification_report"])
        fold_rows.append({
            "fold": fold,
            "checkpoint": str(fold_dir / "resnet50_ema_best.pt"),
            "valid_macro_f1": metrics["macro_f1"],
            "valid_balanced_accuracy": metrics["balanced_accuracy"],
            "valid_accuracy": metrics["accuracy"],
            "valid_roc_auc_ovr_macro": metrics["roc_auc_ovr_macro"],
            "best_epoch": best["epoch"],
            "best_phase": best["phase"],
        })

    summary = pd.DataFrame(fold_rows).sort_values(["valid_macro_f1", "valid_balanced_accuracy"], ascending=False)
    summary.to_csv(out_dir / "cv_summary.csv", index=False)
    print("\nCV summary")
    display(summary)
    best_path = Path(summary.iloc[0]["checkpoint"])
    print("selected best checkpoint by validation macro-F1:", best_path)
    return out_dir, summary, best_path

# Run training when ready. This is intentionally not auto-executed.
# cv_out_dir, cv_summary, best_path = train_resnet50_stratified_cv()
````
